# Analysing The Data

## Question 1
What are the top 10 most populated countries in 2022?
   - How has the population of these countries changed over the years?


The top 10 most populated countries are listed in my sorted DataFrame.They are:

China (CHN)
India (IND)
United States (USA)
Indonesia (IDN)
Pakistan (PAK)
Nigeria (NGA)
Brazil (BRA)
Bangladesh (BGD)
Russia (RUS)
Mexico (MEX)


```python
# To analyze how the population of these countries has changed over the years,
# I will create a separate DataFrame containing only the data for these countries and 
# plot the population trends over time.

# Selecting top 10 most populated countries
top_10_countries = df.sort_values(by="2022 Population", ascending=False).head(10)

# Extracting only population columns for the selected countries
pop_columns = ['Country', '1970 Population', '1980 Population', '1990 Population', 
               '2000 Population', '2010 Population', '2015 Population', 
               '2020 Population', '2022 Population']
top_10_populations = top_10_countries[pop_columns]

# Transposing dataframe for easier plotting
top_10_populations = top_10_populations.set_index('Country').transpose()

# Plotting
plt.figure(figsize=(15, 8))
for country in top_10_populations.columns:
    plt.plot(top_10_populations.index, top_10_populations[country], label=country)

plt.title('Population Trend of Top 10 Most Populated Countries (1970-2022)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Question 2
Is there any correlation between population size and other variables like area, density, etc.?
   - How does the population growth rate correlate with other factors?

First I will look at the correlation between population size and other variables like area and density!

```python
correlation_population = df[['2022 Population', 'Area (km²)', 'Density (per km²)']].corr()
correlation_population

```

Correlation between population growth rate and other factors

```python
correlation_growth_rate = df[['Growth Rate', 'Area (km²)', 'Density (per km²)']].corr()
correlation_growth_rate
```

```python
#visualization using heatmaps
plt.figure(figsize=(10, 8))

plt.subplot(1, 2, 1)
sns.heatmap(correlation_population, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix: Population, Area, Density')

plt.subplot(1, 2, 2)
sns.heatmap(correlation_growth_rate, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix: Growth Rate, Area, Density')

plt.tight_layout()
plt.show()
```

## Question 4
What is the average growth rate of population for different continents?
   - Are there any continents experiencing negative population growth?

```python
# Calculating the average growth rate for each continent
average_growth_rate = df.groupby('Continent')['Growth Rate'].mean()

print("Average Growth Rate for Different Continents:")
print(average_growth_rate)

# Identify continents with negative population growth
continents_with_negative_growth = average_growth_rate[average_growth_rate < 1]
if continents_with_negative_growth.empty:
    print("No continents are experiencing negative population growth.")
else:
    print("Continents experiencing negative population growth:")
    print(continents_with_negative_growth)
```

## Question 5
Can you predict the future population of countries/continents based on the data and growth rates?
    - What factors might influence population changes in the future?

To predict future populations of continents based on the data and growth rates, I will use a mathematical model known as the exponential growth model. The formula for exponential growth is:

$P(t)=P_0 * (1+r)^t$




```python

# Defining a function to predict future population
def predict_population(initial_population, growth_rate, years):
    return initial_population * (1 + growth_rate) ** years

selected_continent = input('Enter continent of choice:')
selected_year = 2022

# Filtering data for the selected continent
continent_data = df[df['Continent'] == selected_continent]

# Calculating average growth rate for the continent
avg_growth_rate = continent_data['Growth Rate'].mean()

# Get the initial population for the selected year
initial_population = continent_data[f'{selected_year} Population'].mean()

# Predict population for future years
future_years = [2025, 2030, 2035]  
predicted_populations = [predict_population(initial_population, avg_growth_rate, year - selected_year) for year in future_years]


for year, population in zip(future_years, predicted_populations):
    print(f"Predicted population for {selected_continent} in {year}: {int(population)}")
```
